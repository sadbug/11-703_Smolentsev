```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mp
```

```python
file = r'../Homework1/train.csv'
data = pd.read_csv(file)
```

```python
data.info()
```

```python
mp.hist(x = [data[data['Survived']==1]['Sex'],data[data['Survived']==0]['Sex']],stacked=True, color = ['black','blue'])
mp.title("Зависимость количества выживших от пола")
```

```python
mp.hist(x = [data[data['Survived']==1]['Age'],data[data['Survived']==0]['Age']],stacked=True, color = ['black','blue'])
mp.title("Зависимость количества выживших от возраста")
```

```python
mp.hist(x = [data[data['Survived']==1]['Pclass'],data[data['Survived']==0]['Pclass']],stacked=True, color = ['black','blue'])
mp.title("Зависимость количества виживших от класса")
```
